```python
import pandas as pd
df = pd.read_csv("../cleaned_water_data.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#correlation matrix
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.savefig("heatmap.png", bbox_inches='tight')
plt.show()
```

```python
#histogarm
df.hist(figsize=(12,10))
plt.savefig("hist.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Boxplot of Water Quality Features")
plt.savefig("boxplot.png", bbox_inches='tight')
plt.show()
```

```python
mean_values = df[['pH', 'Iron', 'Nitrate', 'Chloride',
                  'Sulfate', 'Conductivity']].mean()

plt.figure(figsize=(10,5))
mean_values.plot(kind='bar')

plt.title("Average Values of Water Quality Parameters")
plt.xlabel("Parameters")
plt.ylabel("Average Value")
plt.xticks(rotation=45)
plt.savefig("bar.png", bbox_inches='tight')
plt.show()
```

```python
sns.pairplot(df.sample(100)[['pH','Iron','Nitrate','Chloride','Potability']])
plt.savefig("pair.png", bbox_inches='tight')
plt.show()
```
